```python
X= rate.drop(columns='BaseRate')
y= rate['BaseRate']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2)
```

```python
##from sklearn.linear_model import LinearRegression
##from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['CashReserveRatio','CashDepositRatio','MarginalFacilityRate','TermDepositRate','CallMoneyRate','TreasuryBill','ReverseRepoRate','BankRate', 'SavingsDepositRate','Inflation', 'RepoRate']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['CashReserveRatio','CashDepositRatio','MarginalFacilityRate','TermDepositRate','CallMoneyRate','TreasuryBill','ReverseRepoRate','BankRate', 'SavingsDepositRate','Inflation', 'RepoRate']),remainder= 'passthrough')
```

```python
lr=RandomForestRegressor()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred= pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=i)
    lr=RandomForestRegressor()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
